```python
import pandas as pd
```

```python
#Import demographics dataset
file_path = "C:/Users/VivoBook/Desktop/Classwork/capstoneproject-2022/Data/demographics.csv"
demo_df = pd.read_csv(file_path)
demo_df.head()
```

```python
# Drop unnecessary columns
demo_df = demo_df.drop(labels=["id. Response ID", "Unnamed: 16", "Unnamed: 17",
                               "Unnamed: 18", "Unnamed: 19", "Unnamed: 20", "Unnamed: 21", "Unnamed: 22"], axis=1)
demo_df.head()
```

```python
# Drop NaNs
demo_df = demo_df.dropna()
demo_df.head()
```

```python
# Concatenate Previous EV columns into one
cols_concat = ["Previous Plug in HEVs", "Previous BatteryEVs", "Previous HybridEVs"]

# Convert them to type str
demo_df[cols_concat] = demo_df[cols_concat].astype('str')

# Then concatenate them as follows
demo_df["previous_evs"] = demo_df[cols_concat].T.agg('_'.join)
demo_df.head()
```

```python
# Drop old Previous EV columns and number of people
demo_df = demo_df.drop(labels=["Previous Plug in HEVs", "Previous BatteryEVs", "Previous HybridEVs", "Number of people in the household"], axis=1)
demo_df.head()
```

```python
# Fix previous_evs values to 0 for no and 1 for yes
demo_df.loc[demo_df["previous_evs"] == "0.0_0.0_0.0", "previous_evs"] = "0"  
demo_df.loc[demo_df["previous_evs"] != "0.0_0.0_0.0", "previous_evs"] = "1"  
demo_df.head()
```

```python
# Rename and re-order columns to match database table and types
demo_df.columns = ["year", "car_main", "household_income",
                   "importance_of_gge", "homeowner", "education",
                   "commute_distance", "age", "gender", "number_of_vehicles",
                   "annual_miles", "previous_evs"]

column_to_move = demo_df.pop("previous_evs")
demo_df.insert(2, "previous_evs", column_to_move)

demo_df = demo_df.astype({"year": "int", "previous_evs": "int", "household_income": "float",
                          "homeowner": "int", "education": "int", "age": "int", "gender": "int"})
demo_df.dtypes
```

```python
# Create sample dataframe and save as csv
demo_sample_df = demo_df.sample(frac=0.5)
demo_sample_df.to_csv("demo_db.csv", index=False)
```
