```python
#Import incentives dataset
file_path = "C:/Users/VivoBook/Desktop/Classwork/capstoneproject-2022/Database/incentives_db.csv"
incentives_df = pd.read_csv(file_path)
incentives_df.head()
```

```python
# Convert to datetime
incentives_df["start_date"] = pd.DatetimeIndex(pd.to_datetime(incentives_df["start_date"])).year
incentives_df["end_date"] = pd.DatetimeIndex(pd.to_datetime(incentives_df["end_date"])).year
incentives_df.head()
```

```python
#Import sales_pop dataset, change year to int
file_path = "C:/Users/VivoBook/Desktop/Classwork/capstoneproject-2022/Database/sales_pop.csv"
sales_pop = pd.read_csv(file_path)
sales_pop["incentive_count"] = 0
# sales_pop["year"] = sales_pop["year"].astype("int")
sales_pop.head()
```

```python
# for loop iterating for years 2015-2022, count incentives per county
# (columns: county, year, incentive_count)
def county_year_incentives(county):
    chosen_county = incentives_df.loc[incentives_df["county"] == county]
    for year in range(2015,2023):
        incentive_count = 0
        for index, row in chosen_county.iterrows():
            if row["start_date"] <= year and row["end_date"] >= year:
                incentive_count = incentive_count + 1
                sales_pop.loc[(sales_pop["county"] == county) & (sales_pop["year"] == year),["incentive_count"]] = incentive_count
```
