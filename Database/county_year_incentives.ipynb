```python
# for loop iterating for years 2015-2022, count incentives per county
# (columns: county, year, incentive_count)
def county_year_incentives(county):
    chosen_county = incentives_df.loc[incentives_df["county"] == county]
    for year in range(2015,2023):
        incentive_count = 0
        for index, row in chosen_county.iterrows():
            if row["start_date"] <= year and row["end_date"] >= year:
                incentive_count = incentive_count + 1
                sales_pop.loc[(sales_pop["county"] == county) & (sales_pop["year"] == year),["county_count"]] = incentive_count
```

```python
# for loop for state wide incentives, add count to each county for relevant years
def statewide_incentives(county):
    statewide = incentives_df.loc[incentives_df["county"] == "All"]
    for year in range(2015,2023):
        statewide_count = 0
        for index, row in statewide.iterrows():
            if row["start_date"] <= year and row["end_date"] >= year:
                statewide_count = statewide_count + 1
                sales_pop.loc[(sales_pop["year"] == year),["state_count"]] = statewide_count
```

```python
# for loop iterating through max_amount, add non-customer incentives and max customer incentive
# (columns: county, year, incentive_count, max_amount)
def county_year_max(county):
    chosen_county = incentives_df.loc[incentives_df["county"] == county]
    for year in range(2015,2023):
        # sub-dataframe for year based on chosen_county
        chosen_year = chosen_county.loc[(chosen_county["start_date"] <= year) & (chosen_county["end_date"] >= year)]
        
        # Max EV Charger incentive
        charger_max = chosen_year[(chosen_year["ev_type"]=="EV Charger")]["max_amount"].max()
        if np.isnan(charger_max):
            charger_max = 0
        
        # Non-customer incentives summed with EV Charger        
        non_customer_max = chosen_year[(chosen_year["ev_type"]!="EV Charger") & (chosen_year["customer"]==0)]["max_amount"].sum()
        sales_pop.loc[(sales_pop["county"] == county) & (sales_pop["year"] == year),["county_max_nc"]] = non_customer_max + charger_max
                
        # Max customer incentive
        customer_max = chosen_year[(chosen_year["ev_type"]!="EV Charger") & (chosen_year["customer"]==1)]["max_amount"].max()
        if np.isnan(customer_max):
            customer_max = 0
        sales_pop.loc[(sales_pop["county"] == county) & (sales_pop["year"] == year),["county_max_c"]] = customer_max
```

```python
# For loop CA wide incentive max_amount
def CA_max(county):
    statewide = incentives_df.loc[incentives_df["county"] == "All"]
    for year in range(2015,2023):
        # sub-dataframe for year based on statewide
        chosen_year = statewide.loc[(statewide["start_date"] <= year) & (statewide["end_date"] >= year)]
        
        # Non-customer incentives summed with EV Charger        
        state_max = chosen_year["max_amount"].sum()
        sales_pop.loc[(sales_pop["year"] == year),["state_max"]] = state_max
```

```python
# Run CA_max for 2015-2023
CA_max("All")
sales_pop.iloc[48:100]
```

```python
# Add total_incentives, county_max, total_max columns. Drop county_max_nc and county_max_c
sales_pop.insert(8, "total_incentives", (sales_pop["county_count"] + sales_pop["state_count"]))
sales_pop.insert(11, "county_max", (sales_pop["county_max_nc"] + sales_pop["county_max_c"]))
sales_pop.insert(13, "total_max", (sales_pop["county_max"] + sales_pop["state_max"]))
sales_pop = sales_pop.drop(columns=["county_max_nc", "county_max_c"])
```

```python
# Drop NaNs, change dtypes and column names
sales_pop = sales_pop.sort_values(by=["county", "year"], axis=0, ascending=True).reset_index(level=None, drop=True, inplace=False).dropna()
sales_pop = sales_pop.astype({"year": "int", "number_sales": "int"})
sales_pop.head()
```

```python
# Save sales_pop as county_year_merged
sales_pop.to_csv("county_year_merged.csv", index=False)
```
